Задание 1\
a) $f(x) = x^2 -30x + 1$\
**Теорема** Если функция $y = f(x)$ имеет экстремум в точке $x = x_0$, то в этой точке производная фцнкции либо равна нулю, либо не существуетю.



Приравняем производную функции к нулю и определим критические точки.
$f'(x) = 2x - 30 = 0$\
$x = 15$\
$f''(x) > 0$ ==> график выпуклый вниз, у нас локальный минимум.\
Получется наша функция находится на интервале: $(- \infty; 15) \cup (15; + \infty)$  со следующими знаками на каждом интервале: (-) и (+)\
Ответ: функция убывает при $x \in (- \infty; 15)$, возрастает при $x \in (15; + \infty)$, 15 - экстремум, точка минимума функции.

b) $g(x) = -x^3 + 5x - 7x^2 + 17$\
$g'(x) = -3x^2 - 14x + 5 = 0$\
$D = 14^2 - 4*(-3)*5 = 196 + 60 = 256 = 16^2$\
$x_1 = \frac{14 + 16}{-6} = -5$\
$x_2 = \frac{14 - 16}{-6} = \frac{1}{3}$\
$g''(x) = -6x - 14$. \
При $x \in (- \infty; -5)$ $g''(x) > 0$  ==> график выпуклый вниз, у нас локальный минимум.\
При $x \in (-5; \frac{1}{3})$ $g''(x) < 0$  ==> график выпуклый вверх, у нас локальный максимум.\
Получется наша функция находится на интервале: $(- \infty; -5) \cup (-5; \frac{1}{3}) \cup (\frac{1}{3}; +\infty)$  со следующими знаками на каждом интервале: (-) и (+) и (-).\
Ответ: функция убывает при $x \in (- \infty; -5) \cup (\frac{1}{3}; +\infty)$, возрастает при $x \in (-5; \frac{1}{3})$, -5 - экстремум, точка локального минимума функции, $\frac{1}{3} - экстремум, точка локального максимума функции$.

c) $h(x) = x^2 - \frac{1}{x}$\
$h'(x) = 2x + \frac{1}{x^2} = 0$\
$x_1 = -\frac{1}{\sqrt[3]{2}}$\
$x_2 = -\frac{1}{2\sqrt[3]{2}} - \frac{\sqrt{3}}{2\sqrt[3]{2}}i$\
$x_3 = \frac{1}{2\sqrt[3]{2}} + \frac{\sqrt{3}}{2\sqrt[3]{2}}i$\
$h''(x_1) = 2(1-\frac{1}{x^3}) > 0$ ==> график выпуклый вниз, у нас локальный минимум.


d) $j(x) = \frac{1}{x^5} + x^{17} + 8$\
$j'(x) = \frac{17 x^{22} - 5}{x^6} = 0$\
$x_1 = \sqrt[22]{\frac{5}{17}} = 0.946$\
$x_2 = -\sqrt[22]{\frac{5}{17}} = -0.946$\
$j''(x_1) > 0$ ==> у нас локальный минимум.\
$j''(x_2) < 0$ ==> у нас локальный максимум.

В данной задаче я искала экстремумы только для действительных чисел, так как вычисление второй производной для комплексного аргумента может не иметь физического смысла в контексте нахождения экстремума, так как нет понятия "локального максимума" или "локального минимума" для функций, принимающих комплексные значения.

```python
import sympy as sp
```

```python
def extreme(expr, var):
    derivative = sp.diff(expr, var)
    critical_points = sp.solve(derivative, var)
    derivative2 = sp.diff(derivative, var)
    res = [(point, derivative2.subs({var: point})) for point in critical_points if point.is_real]
    return res
```

```python
x = sp.symbols('x')
f1 = x**2 - 30 * x + 1
f2 = -x**3 + 5*x - 7 * x**2 + 17
f3 = x**2 - 1/x
f4 = 1/x**5 + x**17 + 8
```

```python
functions = [f1, f2, f3, f4]
for i in range(len(functions)):
    print(f'Функция f{i+1} = {functions[i]}')
    for point, value in extreme(functions[i], x):
        if value > 0:
            print(f"Найден экстремум в точке x = {point}, у нас локальный минимум.")
        else:
            print(f"Найден экстремум в точке x = {point}, у нас локальный максимум.")
    print('\n')
```

Найдем точки перегиба этих функций в Python.

```python
def inflection_points(expr, var):
    derivative2 = sp.diff(expr, var, 2)
    critical_points = sp.solve(derivative2, var)
    return [point for point in critical_points if point.is_real]
```

```python
for i in range(len(functions)):
    print(f'Функция f{i+1} = {functions[i]}')
    inflection_point =  inflection_points(functions[i], x)
    if inflection_point:
        for point in inflection_point:
            print(f'Точка перегиба x = {point}')
    else:
        print('Точки перегиба нет!')
    print('\n')
```

Модифицируем функцию, осуществляющую метод градиентного спуска

trace в данном коде - это история значений, но я ее закомментировала тк долго работало.

```python
def gradient_descent(f, x_start, learning_rate=0.01, epsilon=1e-6, num_iterations=1000, delta=1e-3):
    x_curr = x_start
    df_x = sp.diff(f, x)

    # trace = []
    # trace.append(x_curr)

    for i in range(num_iterations):
        x_new = x_curr - learning_rate * df_x.subs(x, x_curr)
        # trace.append(round(x_new, 2))

        if abs(x_new - x_curr) < epsilon or abs(df_x.subs(x, x_new)) < delta:
            return x_curr

        x_curr = x_new

    return x_curr
```

```python
def beautiful_graph(functions):
    x_vals = np.array([x for x in np.arange(-1000, 1000, 0.1)])
    fig, axes = plt.subplots(nrows=(4 // 2) + (4 % 2), ncols=2, figsize=(17, (4 // 2 + 1) * 2))

    # Разные пределы для каждой функции
    y_limits = [(-500, 1000), (-80, 80), (-40, 40), (-150, 150)]
    x_limits = [(-200, 200), (-20, 20), (-10, 10), (-5, 5)]

    # Градиентный спуск
    learning_rate = 0.01
    epsilon = 1e-6
    num_iterations = 1000
    delta = 1e-3

    # Построение графиков
    for i, (func, ax) in tqdm(enumerate(zip(functions, axes.flatten()))):
        f_lambdified = sp.lambdify(x, func, modules=["numpy"])
        y_vals = f_lambdified(x_vals)

        # Построение графика функции
        ax.plot(x_vals, y_vals, label=f'Функция f{i+1}', color='blue')

        x_starts = [random.randint(-5, 5) for _ in range(5)]
        x_end = []
        for x_start in x_starts:
            gradient = gradient_descent(func, x_start)
            x_end.append(gradient)

        # Определяем минимум из окончательных значений
        minimum_x = min(x_end, key=lambda v: f_lambdified(v))
        ax.scatter(float(minimum_x), f_lambdified(float(minimum_x)), color='green', zorder=5, label='Минимальное значение')
        ax.annotate(f'({float(minimum_x):.4f}, {f_lambdified(float(minimum_x)):.4f})',
                    (float(minimum_x), f_lambdified(float(minimum_x))),
                    textcoords='offset points',
                    xytext=(0,10),
                    ha='center', color='green')


        # Настройка заголовка и осей
        ax.set_title(f'График функции $f{i+1}(x) = {sp.latex(func)}$ с наименьшим локальным минимумом')
        ax.set_xlabel('x')
        ax.set_ylabel(f'f{i+1}(x)')

        # Установка диапазона по оси Y и X
        ax.set_ylim(y_limits[i])
        ax.set_xlim(x_limits[i])
        ax.grid(True)

        # Выделяем нулевые оси
        ax.axhline(0, color='black', linewidth=1.1)
        ax.axvline(0, color='black', linewidth=1.1)

        ax.legend()

    # Удаление неиспользуемых графиков
    for i in range(len(functions), len(axes.flatten())):
        fig.delaxes(axes.flatten()[i])

    plt.tight_layout()
    plt.show()
```

Я реализую метод Ньютона (метод второго порядка) для поиска минимума функции.
Этот подход использует не только первую, но и вторую производную функции, чтобы быстрее и точнее сходиться к точке экстремума.

Для понимания алгоритма я опираюсь на описание из англоязычной статьи про Newton’s method in optimization, а также на материал на русском языке, где подробно разобрано, как метод Ньютона применяется именно для задач минимизации.

```python
def newton_method(func, x_start, learning_rate=1.0, epsilon=1e-6, num_iterations=100):
    x = sp.symbols('x')
    gradient = sp.diff(func, x)
    hessian = sp.diff(gradient, x)

    x_curr = x_start
    for i in range(num_iterations):
        grad_val = gradient.subs(x, x_curr)
        hessian_val = hessian.subs(x, x_curr)

        x_new = x_curr - learning_rate * (grad_val / hessian_val)

        if abs(x_new - x_curr) < epsilon or abs(grad_val) < epsilon:
            break

        x_curr = x_new

    return x_curr
```
